**Importing necessary Libraries**

```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from urllib.request import urlopen
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
html = urlopen(url) 
soup = BeautifulSoup(html, 'html.parser')
import re
```

**Web scraping using Beautiful soup**

```python


postal_code = []
borough = []
neighborhood = []

data_dict = {}
tables = soup.find_all('table',{'class':'wikitable sortable'})
for table in tables:
    if len(table) > 1:
        rows = table.find_all('tr')
        
        for row in rows[1:]:
            cells = row.find_all('td')
            if len(cells)>0:
                # for postal code list
                # using regular expressions sub
                p_code = cells[0]
                p_code = re.sub('<[^>]*>', '', str(p_code))
                postal_code.append(p_code)
   
                # for borough list
                burro = cells[1]
                burro = re.sub('<[^>]*>', '', str(burro))
                borough.append(burro)
                
                # neighbourhood list
                ngb = cells[2]
                ngb = re.sub('<[^>]*>', '', str(ngb))
                neighborhood.append(ngb)

                
                
data_dict['Postal Code'] = [a.strip() for a in postal_code]# removing '\n' from the dtrings
data_dict['Borough'] = [a.strip() for a in borough]
data_dict['Neighborhood'] = [a.strip() for a in neighborhood]
```

**Converting Dict to **

```python
df = pd.DataFrame(data_dict)
```

**removing Not assigned Buroughs**

```python
df_filtered = df[df['Borough'] != 'Not assigned']
df_filtered
```
